```python
# To perform our task we will import the below libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Importing the "insurance.csv" file to perform our analysis, our_data.head() will help us better understand the given values

our_data = pd.read_csv("insurance.csv")
our_data.head()
```

We first filter and sort the dataset by age in ascending order, which provides insight for better understanding the relationship between charges and age. By sorting the data, we can more clearly observe any patterns or trends. We then proceed with the rest of the code, which includes fitting a linear regression model, making predictions, and plotting the scatter plot with the best-fit line.

```python
# Filter our_data and sort value by charges

filtered_data = our_data.sort_values(by='charges')
filtered_data.head() # Display the first 5 observations
```

```python
# Filter and sort the data
filtered_data = our_data.sort_values(by='charges')
filtered_data.tail() # Display the last 5 observations
```

```python
# The iloc function allows use to select rows/columns from our_data for plotting 
# x selects all rows and column 1 indexing from left to right = "age"
# y selects all rows indexing from right to left -1 will = to "charges"

x = our_data.iloc[:,:1].values
y = our_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('age')
plt.ylabel('charges')
plt.title('Age vs Charges')
plt.show()
```

```python
# Fit the model
regression = LinearRegression()
regression.fit(x,y)
```

```python
# Plot the data and model
predictions = regression.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,predictions,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('How Age Affects Insurance Cost')
plt.show()
```
